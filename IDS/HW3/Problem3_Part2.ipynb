```python
# import some of the needed libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the life expectancy data into a dataframe

url = "https://raw.githubusercontent.com/bforoura/IDS/main/HW3/Life_Expectancy_Mac.csv"

lifeExpectancyDF = pd.read_csv(url, sep=",")
```

```python
# Display the current dataframe

lifeExpectancyDF
```

```python
# Delete any rows with population of 0

lifeExpectancyDF = lifeExpectancyDF.dropna(subset=['Population'])
```

```python
# Fill in any other missing values with the averages of the respective columns

lifeExpectancyDF['Life_Expectancy '] = lifeExpectancyDF['Life_Expectancy '].fillna(lifeExpectancyDF['Life_Expectancy '].mean())

lifeExpectancyDF['Adult_Mortality'] = lifeExpectancyDF['Adult_Mortality'].fillna(lifeExpectancyDF['Adult_Mortality'].mean())

lifeExpectancyDF['Alcohol'] = lifeExpectancyDF['Alcohol'].fillna(lifeExpectancyDF['Alcohol'].mean())

lifeExpectancyDF['Percentage_Expenditure'] = lifeExpectancyDF['Percentage_Expenditure'].fillna(lifeExpectancyDF['Percentage_Expenditure'].mean())

lifeExpectancyDF[' BMI '] = lifeExpectancyDF[' BMI '].fillna(lifeExpectancyDF[' BMI '].mean())

lifeExpectancyDF['Total_Expenditure'] = lifeExpectancyDF['Total_Expenditure'].fillna(lifeExpectancyDF['Total_Expenditure'].mean())

lifeExpectancyDF['GDP'] = lifeExpectancyDF['GDP'].fillna(lifeExpectancyDF['GDP'].mean())

lifeExpectancyDF['Population'] = lifeExpectancyDF['Population'].fillna(lifeExpectancyDF['Population'].mean())

lifeExpectancyDF['Schooling'] = lifeExpectancyDF['Schooling'].fillna(lifeExpectancyDF['Schooling'].mean())
```

```python
# Display the cleansed dataframe

lifeExpectancyDF
```

```python
# How does Adult mortality rates affect life expectancy?

lifeExpectancyDF['Adult_Mortality'].corr(lifeExpectancyDF['Life_Expectancy '])
```

```python
# Does life expectancy have positive or negative correlation with eating habits?

lifeExpectancyDF[' BMI '].corr(lifeExpectancyDF['Life_Expectancy '])
```

```python
# Does life expectancy have positive or negative correlation with drinking alcohol?

lifeExpectancyDF['Alcohol'].corr(lifeExpectancyDF['Life_Expectancy '])
```

```python
# Does life expectancy have positive or negative correlation with social factors?

populationCorr = lifeExpectancyDF['Population'].corr(lifeExpectancyDF['Life_Expectancy '])

print(f"Correlation between Population and Life Expectancy: {populationCorr}")
```

```python
# Does life expectancy have positive or negative correlation with economic factors?


percentageExpenditureCorr = lifeExpectancyDF['Percentage_Expenditure'].corr(lifeExpectancyDF['Life_Expectancy '])
totalExpenditureCorr = lifeExpectancyDF['Total_Expenditure'].corr(lifeExpectancyDF['Life_Expectancy '])
gdpCorr = lifeExpectancyDF['GDP'].corr(lifeExpectancyDF['Life_Expectancy '])

print(f"Correlation between Percentage Expenditure and Life Expectancy: {percentageExpenditureCorr}")
print(f"Correlation between Total Expenditure and Life Expectancy: {totalExpenditureCorr}")
print(f"Correlation between GDP and Life Expectancy: {gdpCorr}")
```

```python
# What is the impact of schooling on the lifespan of humans?

lifeExpectancyDF['Schooling'].corr(lifeExpectancyDF['Life_Expectancy '])
```

```python
# Create and plot several linear regression models for life expectancy
# (as the dependent variable) and the independent variables that had the highest correlation
# with life expectancy as found in your correlation analysis.

# Independent variables with highest correlation with life expectancy: Schooling, BMI, GDP (0.796, 0.522, 0.466 respectively)

# Drawing scotterplot of Schooling and Life Expectancy

lifeExpectancyDF.plot.scatter(x='Schooling', y='Life_Expectancy ', color="DarkGreen")
```

```python
# Regression line of Schooling and Life Expectancy

from sklearn.linear_model import LinearRegression

# Save the schooling column in X
X = lifeExpectancyDF[['Schooling']]

# Save the Life_Expectancy column in Y
Y = lifeExpectancyDF[['Life_Expectancy ']]

# Create the model
schooling_LRmodel = LinearRegression()

# Perform linear regression
schooling_LRmodel.fit(X, Y)

# Make predictions
Y_pred = schooling_LRmodel.predict(X)


# Visualize the dataset and the regression line:
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')

plt.show()
```

```python
# Drawing scotterplot of BMI and Life Expectancy

lifeExpectancyDF.plot.scatter(x=' BMI ', y='Life_Expectancy ', color="DarkGreen")
```

```python
# Regression line of BMI and Life Expectancy

from sklearn.linear_model import LinearRegression

# Save the BMI column in X
X = lifeExpectancyDF[[' BMI ']]

# Save the Life_Expectancy column in Y
Y = lifeExpectancyDF[['Life_Expectancy ']]

# Create the model
BMI_LRmodel = LinearRegression()

# Perform linear regression
BMI_LRmodel.fit(X, Y)

# Make predictions
Y_pred = BMI_LRmodel.predict(X)


# Visualize the dataset and the regression line:
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')

plt.show()
```

```python
# Drawing scotterplot of GDP and Life Expectancy

lifeExpectancyDF.plot.scatter(x='GDP', y='Life_Expectancy ', color="DarkGreen")
```

```python
# Regression line of GDP and Life Expectancy

from sklearn.linear_model import LinearRegression

# Save the GDP column in X
X = lifeExpectancyDF[['GDP']]

# Save the Life_Expectancy column in Y
Y = lifeExpectancyDF[['Life_Expectancy ']]

# Create the model
GDP_LRmodel = LinearRegression()

# Perform linear regression
GDP_LRmodel.fit(X, Y)

# Make predictions
Y_pred = GDP_LRmodel.predict(X)


# Visualize the dataset and the regression line:
plt.scatter(X, Y)
plt.plot(X, Y_pred, color='red')

plt.show()
```

```python
# Provide the equation of each model

# The equation of the line of Schooling and Life Expectancy:

print("The Schooling slope: ", schooling_LRmodel.coef_)
print("The Schooling intercept: ", schooling_LRmodel.intercept_)

# The equation of the line of BMI and Life Expectancy:

print("The BMI slope: ", BMI_LRmodel.coef_)
print("The BMI intercept: ", BMI_LRmodel.intercept_)

# The equation of the line of GDP and Life Expectancy:

print("The GDP slope: ", GDP_LRmodel.coef_)
print("The GDP intercept: ", GDP_LRmodel.intercept_)

```

These equations follow slope-intercept form y=m*x+b.

y = the dependent variable

m = slope

x = the independent variable

b = the y intercept

Values are rounded up to four decimal points.

### Schooling and Life Expectancy Equation:
(Life_Expectancy) = 2.3291 * (Schooling) + 41.1573

### BMI and Life Expectancy Equation:
(Life_Expectancy) = 0.2253 * (BMI) + 61.5545

### GDP and Life Expectancy Equation:
(Life_Expectancy) = 0.0003 * (GDP) + 68.3751

```python
# Compute their R2 and MSE values.

# Make predictions using the Schooling, BMI, and GDP columns

y_pred_schooling = schooling_LRmodel.predict(lifeExpectancyDF[['Schooling']])
y_pred_BMI = BMI_LRmodel.predict(lifeExpectancyDF[[' BMI ']])
y_pred_GDP = GDP_LRmodel.predict(lifeExpectancyDF[['GDP']])

# Error analysis

from sklearn.metrics import mean_squared_error, r2_score

print("Schooling MSE: ", mean_squared_error(lifeExpectancyDF[['Life_Expectancy ']], y_pred_schooling))
print("Schooling R2: ", r2_score(lifeExpectancyDF[['Life_Expectancy ']], y_pred_schooling))

print("BMI MSE: ", mean_squared_error(lifeExpectancyDF[['Life_Expectancy ']], y_pred_BMI))
print("BMI R2: ", r2_score(lifeExpectancyDF[['Life_Expectancy ']], y_pred_BMI))

print("GDP MSE: ", mean_squared_error(lifeExpectancyDF[['Life_Expectancy ']], y_pred_GDP))
print("GDP R2: ", r2_score(lifeExpectancyDF[['Life_Expectancy ']], y_pred_GDP))
```
